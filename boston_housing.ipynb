```python
import numpy as np 
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
from pandas.plotting import scatter_matrix
```

```python
housing_data = load_boston()
```

```python
housing_data 
```

```python
housing_df = pd.DataFrame(housing_data.data, columns = housing_data.feature_names)
housing_df['MEDV'] = housing_data.target
housing_df
```

```python
housing_df = housing_df.rename(columns={"CRIM": "crime", "ZN": "Zone_Proportion", "INDUS": "Proportion_of_Non-Retail",
                                      "CHAS": "Charles_River", "NOX": "Nitric_Oxide", 
                                      "RM": "Average_Room", "AGE": "Unit_Age",
                                        "DIS": "distances_to_emp_centre",
                                         "RAD": "Highway_Access_Index",
                                         "TAX": "property_tax_rate",
                                         "PTRATIO": "pupil_teacher_ratio",
                                        "B": "Blacks_by_Town",
                                        "LSTAT": "Lower_Status_%",
                                        "MEDV": "Median_Home_Value"})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
housing_df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# from pandas.tools.plotting import scatter_matrix # For older versions of Pandas

attributes = ["Median_Home_Value", "crime", "Zone_Proportion", "Proportion_of_Non-Retail", "Charles_River", 
              "Nitric_Oxide", "Average_Room", "Unit_Age", "distances_to_emp_centre", "Highway_Access_Index",
              "property_tax_rate", "pupil_teacher_ratio", "Blacks_by_Town", "Lower_Status_%"]     
             
scatter_matrix(housing_df[attributes], figsize=(30, 30));
```
